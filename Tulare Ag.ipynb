## Changing the wide form of the data to long form in order to visualize it

```python
df3 = df2.melt(id_vars='index', var_name='crop', value_name='amount')
df3
```

```python
alt.Chart(df3, title='What is the most produced crop?').mark_area().encode(
    x='index:N',
    y='amount:Q').facet(facet='crop:N',columns=2)
```

```python
df3.to_csv("crops.csv", index=False)
```
